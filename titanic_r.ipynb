```python
import pandas as pd
import numpy as np
import math as ma
```

```python
data=pd.read_csv(r"titanic_x_y_train.csv",delimiter=",")
ta=data.copy()
```

```python
#a=ta[ta.Sex=="male"]
#b=ta[ta.Sex=="female"]
#ta.loc[(ta["Age"].isnull()) & (ta.Sex=="male"),"Age"]=a.Age.mean()
#ta.loc[(ta["Age"].isnull()) & (ta.Sex=="female"),"Age"]=a.Age.mean()


```

```python
def func(s):
    if s=="male":
        
        
        return 1
    else:
        return 2


ta["Gender"]=ta.Sex.apply(func)
del ta["Sex"]
del ta["Name"]
```

```python
def func1(s):
    if s=="S":
        
        
        return 1
    if s=="Q":
        
        return 2
    else:
        return 3
ta["emb"]=ta.Embarked.apply(func1)   
```

```python
ta.drop(["Ticket","Cabin","Fare","Embarked"], axis = 1,inplace=True)
```

```python
#a=ta[ta.Gender==1]
#b=ta[ta.Gender==2]
a=ta.loc[(ta["Survived"]==1) & (ta.Gender==1)]
b=ta.loc[(ta["Survived"]==1) & (ta.Gender==2)]
c=ta.loc[(ta["Survived"]==0) & (ta.Gender==2)]
d=ta.loc[(ta["Survived"]==0) & (ta.Gender==2)]

ta.loc[(ta["Age"].isnull()) & (ta.Gender==1) & (ta.Survived==1),"Age"]=a.Age.mean()
ta.loc[(ta["Age"].isnull()) & (ta.Gender==2) & (ta.Survived==1),"Age"]=b.Age.mean()
ta.loc[(ta["Age"].isnull()) & (ta.Gender==2) & (ta.Survived==0),"Age"]=c.Age.mean()
ta.loc[(ta["Age"].isnull()) & (ta.Gender==1) & (ta.Survived==0),"Age"]=d.Age.mean()
#ta.loc[(ta["Age"].isnull()) & (ta.Gender==2),"Age"]=b.Age.mean()

















```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
y=ta.Survived
x=ta.drop(["Survived"],axis=1)
clf.fit(x,y)
clf.score(x,y)
```
